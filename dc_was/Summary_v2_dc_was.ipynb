### Volume-to-Count Ratio after Analyster

#### Low Volume Counts $(<5000)$

```python
sel = (vcol_data['Count']>0) & (vcol_data['Count']<=5000)
vcol_data[['Count','Volume']][sel].plot(kind='Scatter', 
                                        x='Count', y='Volume', 
                                        alpha=0.5, color = 'blue',
                                        xlim=(0,8000), ylim=(0,8000),
                                        figsize=(8,6))
```

#### Medium Volume Counts  $(5000-15000)$

```python
sel = (vcol_data['Count']>5000) & (vcol_data['Count']<=15000)
vcol_data[['Count','Volume']][sel].plot(kind='Scatter', 
                                        x='Count', y='Volume', 
                                        alpha=0.5, color = 'green',
                                        xlim=(0,20000), ylim=(0,20000),
                                        figsize=(8,6))
```

#### High Volume Counts $(>15000)$

```python
sel = (vcol_data['Count']>15000) #& (vcol_data['Count']<=200000)
vcol_data[['Count','Volume']][sel].plot(kind='Scatter', 
                                        x='Count', y='Volume',
                                        alpha=0.5, color = 'red',
                                        figsize=(8,6))
```
